# Mutual Fund Analytics - Exploratory Data Analysis (EDA)

## Introduction

#### Objective

The objective of this Exploratory Data Analysis (EDA) is to understand the mutual fund dataset by analyzing trends, investor behavior, fund performance, and distribution patterns through various visualizations. The analysis helps identify meaningful insights before performing advanced analytics and model development.

---

## Dataset Overview

#### Datasets Used

- Fund Master
- NAV History
- Investor Transactions
- Scheme Performance
- AUM by Fund House
- Industry Folio Count
- Portfolio Holdings
- Benchmark Indices

#### Tools Used

- Python
- Pandas
- Matplotlib
- Seaborn
- Plotly
- SQLite
- VS Code

---

## NAV Trend Analysis

#### Description

This visualization shows the NAV movement of multiple mutual fund schemes over time from 2022 to 2026. It helps identify long-term growth trends and market fluctuations.

#### Observation

Overall NAV values show an increasing trend with temporary corrections during certain periods.

---

## AUM Growth Analysis

#### Description

This chart compares the Assets Under Management (AUM) across different fund houses over multiple years.

#### Observation
